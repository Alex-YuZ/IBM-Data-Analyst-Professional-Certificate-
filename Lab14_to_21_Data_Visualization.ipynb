## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age as age, WorkWeekHrs as work_week_hours
FROM master

"""

df = pd.read_sql_query(QUERY, conn)
df
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(data=df, x='age', y='work_week_hours', s=150)
plt.xlabel('Age', fontsize=18)
plt.ylabel('Work Week Hours', fontsize=18);
```

**Find the number of respondents who ___only___ work with `MySQL`**.

> ___Step 1:___ Grouping by `Respondent`, calculate the respondent and the number of databases they work with.


> ___Step 2:___ Find the respondents' IDs who work with only one type of database.


> ___Step 3:___ Grouping by `DatabaseWorkedWith`, find the number of respondents who work with `MySQL`.

```python
QUERY = """
SELECT *
FROM DatabaseWorkedWith
"""

df = pd.read_sql_query(QUERY, conn)
df
```

```python
# Step 1
count_db = df.groupby('Respondent').count()
count_db
```

```python
one_db_only = (df.groupby('Respondent').count()==1).sum()
print("{} respondents work with only ONE type of database.".format(one_db_only.values[0]))
```

```python
# Step 2: Filter respondent IDs work with only 1 type of database
one_db_respID = count_db[count_db.DatabaseWorkedWith==1].index
one_db_respID
```

```python
one_db_df = df.loc[df.Respondent.isin(one_db_respID), :]
one_db_df
```

```python
# Step 3
one_db_df.DatabaseWorkedWith.value_counts()
```

So we can see from above, **474** respondents work with `MySQL` only.

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT *
FROM master
WHERE Age BETWEEN 30 AND 35
"""

df = pd.read_sql_query(QUERY, conn)
df.shape
```

```python
wwh = df.groupby(['Age']).WorkWeekHrs.median()
wwh = wwh.to_list()
wwh
```

```python
crh = df.groupby(['Age']).CodeRevHrs.median()
crh = crh.to_list()
crh
```

```python
plt.figure(figsize=(12, 8))
age_range = range(30, 36)
plt.bar(age_range, wwh, label='Work Week Hours', width=.6, color='#66b3ff')
plt.bar(age_range, crh, bottom=wwh, label='Code Review Hours', width=.6, color='#ff9999')

plt.title("Median of WorkWeekHrs and CodeReviewHrs", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel('Hours', fontsize=12)
plt.ylim((0, 55))
plt.yticks(range(0, 60, 5))
plt.legend();
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT *
FROM master
WHERE Age BETWEEN 45 AND 60
"""

df = pd.read_sql_query(QUERY, conn)
df.shape
```

```python
df = df.groupby('Age').ConvertedComp.median().to_frame().reset_index()
df
```

```python
plt.figure(figsize=(12, 8))
plt.plot(df['Age'], df['ConvertedComp'], color='#66b3ff', linewidth=4, marker='o', markersize=11, markerfacecolor='white')

plt.title("ConvertedComp (Median) v.s. Age (45-60 y/o)", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel('Converted Compensation ($)', fontsize=12)
plt.xticks(range(45, 61))
plt.yticks(range(0, 155000, 10000), ["{}K".format(int(i/1000)) for i in range(0, 155000, 10000)]);
```

Plot the median ConvertedComp for all ages from **25 to 30.**

```python
# your code goes here
QUERY = """
SELECT *
FROM master
WHERE Age BETWEEN 25 AND 30
"""

df = pd.read_sql_query(QUERY, conn)
df = df.groupby('Age').ConvertedComp.median().to_frame().reset_index()

plt.figure(figsize=(12, 8))
plt.plot(df['Age'], df['ConvertedComp'], color='#ff9999', linewidth=4, marker='o', markersize=11, markerfacecolor='white')

plt.title("ConvertedComp (Median) v.s. Age (25-30 y/o)", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel('Converted Compensation ($)', fontsize=12)
plt.xticks(range(25, 31))
plt.yticks(range(0, 65000, 5000), ["{}K".format(int(i/1000)) for i in range(0, 65000, 5000)]);
```

**Find the most common DevType**

```python
# your code goes here
QUERY = """
SELECT *
FROM DevType
"""

df = pd.read_sql_query(QUERY, conn)
df.sample(10)
```

```python
df.DevType.value_counts()
```

```python
tmp = df.DevType.str.split(', ').values.flatten()
arrs = []
for i in tmp:
    for j in i:
        arrs.append(j)
        
from collections import Counter
c = Counter(arrs)
c.most_common()
        
```
